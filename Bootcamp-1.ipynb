```python
# Practice Question-2 : Global & local use
i = 30

def outer():
    # Local variable in outer function
    i = 10
    print(i)
    
    def inner():
        # Local variable in inner function
        i = 20
        print(i)
        # nonlocal i   # need to define nonlocal at the start, i cannot be used before defining nonlocal i
        # i = 100
        # print(i)
    
    def inner_global():
        # Modifying the global variable
        global i
        print(i)
        i = 40   # Note: we cannot create a local variable with the same name i
        print(i)
    
    def inner_nonlocal():
        # Modifying the outer function's local variable
        nonlocal i
        print(i)
        i = 50
        print(i)

    inner()  # Call the inner function (local scope)
    inner_global()  # Modify the global variable using global keyword
    inner_nonlocal()  # Modify outer's local variable using nonlocal  

# Call the outer function
outer()

# Print the global variable
print(i)
```

```python
class A:
    def __init__(self):
        super().__init__()
        print("A")
class B:
    def __init__(self):
        print("B")
    
class C(A,B):
    def __init__(self):
        super().__init__()
        print("C")
    
class D(C):
    def __init__(self):
        super().__init__()
        print("D")
class E(C):
    def __init__(self):
        super().__init__()
        print("E")
class F(C):
    def __init__(self):
        print("F")
    
class G(D,E,F):
    def __init__(self):
        super().__init__()
        print("G")

obj = G()
```
